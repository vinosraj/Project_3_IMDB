### Creating and saving files as needed

```python
file_2000 = "Data/final_tmdb_data_2000.csv.gz"

tbdm_2000 = pd.read_csv(file_2000, low_memory=False)

tbdm_2000.head()
```

```python
tbdm_2000.info()
```

```python
file_2001 = "Data/final_tmdb_data_2001.csv.gz"

tbdm_2001 = pd.read_csv(file_2001, low_memory=False)

tbdm_2001.head()
```

```python
tbdm_2001.info()
```

```python
tmdb_results_combined = pd.concat([tbdm_2000,tbdm_2001], ignore_index= True)
```

```python
tmdb_results_combined.info()
```

```python
FOLDER = "Data/"
tmdb_results_combined.to_csv(f"{FOLDER}tmdb_results_combined.csv.gz", compression="gzip", index=False)
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.countplot (data =tmdb_final_eda, x ='certification')
```

```python
What is the average revenue per certification category?
What is the average budget per certification category?
```

```python
tmdb_final_eda.loc[tmdb_final_eda['certification'] == 'R', 'revenue'].mean()
```

```python
Average_Revenue = tmdb_final_eda.groupby(['certification'])['revenue'].mean().sort_values()
Average_Revenue
```

```python
from numpy import mean

sns.barplot(y="revenue",x="certification", data=tmdb_final_eda, 
    estimator= mean,ci=None);
```

```python
sns.set(rc={'figure.figsize':(6,4)})

sorted_nb = tmdb_final_eda.groupby(['certification'])['revenue'].mean().sort_values()
ax= sns.boxplot(x = tmdb_final_eda['certification'], y = tmdb_final_eda['revenue'], order=list(sorted_nb.index))
```

```python
tmdb_final_eda.loc[tmdb_final_eda['certification'] == 'R', 'budget'].mean()
```

```python
Average_Budget = tmdb_final_eda.groupby(['certification'])['budget'].mean().sort_values()
Average_Budget
```

```python
sns.barplot(y="budget",x="certification", data=tmdb_final_eda, 
    estimator= mean,ci=None);
```

```python
sns.set(rc={'figure.figsize':(6,4)})

sorted_nb = tmdb_final_eda.groupby(['certification'])['budget'].mean().sort_values()
sns.boxplot(x = tmdb_final_eda['certification'], y = tmdb_final_eda['budget'], order=list(sorted_nb.index))
```

```python
sns.set(rc={'figure.figsize':(6,4)})

sns.relplot(y="revenue", x="budget", hue="certification", data=tmdb_final_eda);
```
