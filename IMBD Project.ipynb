```python
import pandas as pd
import numpy as np
```

```python
basics_url = "https://datasets.imdbws.com/title.basics.tsv.gz"
ratings_url = "https://datasets.imdbws.com/title.ratings.tsv.gz"
akas_url = "https://datasets.imdbws.com/title.akas.tsv.gz"
```

```python
# load data
basics_df = pd.read_csv(basics_url, sep='\t', low_memory=False)
ratings_df = pd.read_csv(ratings_url, sep='\t', low_memory=False)
akas_df = pd.read_csv(akas_url, sep='\t', low_memory=False)
```

```python
ratings_df.head()
```

```python
akas_df.head()
```

Preposing for Title basics

```python
#replace \n with np.nan 
basics_df = basics_df.replace({'\\N':np.nan})
```

```python
basics_df.head()
```

```python
#eliminate movies that are Nan for runtimeMinutes and genres
basics_df = basics_df.dropna(subset=['runtimeMinutes', 'genres', 'startYear'])
```

```python
#only those with tile movie
basics_df = basics_df.loc[basics['titleType']=='movie']
```

there are nan in startYear so going to drop nan values above

```python
#only movies with start year 2000-2022
basics_df['startYear'] = basics_df['startYear'].astype(int)
basics_df= basics.loc[(basics_df['startYear'] >=2000) & (basics_df['startYear'] <=2022)]
```

```python
#Eliminate documentaries
is_documentary = basics_df['genres'].str.contains('documentary', case=False)
basics_df = basics_df[~is_documentary]
```

```python
basics_df.head()
```

### Preposing for AKAs

keep only US entries.

Replace "\N" with np.nan

```python
akas_df = akas_df.replace({'\\N':np.nan})
```

```python
akas_df.head()
```

```python
#US entries only

akas_df = akas_df.loc[akas_df['region']=='US']
```

### Preposing for Ratings

```python
ratings_df = ratings_df.replace({'\\N':np.nan})
```

```python
ratings_df.head()
```

Filtering one data frame based on another

```python
keepers = basics_df['tconst'].isin(akas_df['titleId'])
basics_df = basics_df[keepers]
basics_df
```
